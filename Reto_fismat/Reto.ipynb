```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el centro y el radio del crculo

# Circulo 1
centro2 = (2, 0)
radio2 = 10

# Circulo 2
centro = (5, 0)
radio = 4

# Generar puntos en el crculo
# Circulo 1
theta = np.linspace(0, 2*np.pi, 100)
x = centro[0] + radio * np.cos(theta)
y = centro[1] + radio * np.sin(theta)

# Circulo 2
theta2 = np.linspace(0, 2*np.pi, 100)
x2 = centro2[0] + radio2 * np.cos(theta2)
y2 = centro2[1] + radio2 * np.sin(theta2)

# Graficar los crculos en la misma figura
plt.figure(figsize=(6, 6))

# Circulo 1
plt.plot(x, y, label='Crculo 1')
plt.plot(centro[0], centro[1], 'bo', label='Centro 1')

# Circulo 2
plt.plot(x2, y2, label='Crculo 2')
plt.plot(centro2[0], centro2[1], 'ro', label='Centro 2')

# Configurar ejes y etiquetas
plt.axis('equal')
plt.xlabel('Imaginarios')
plt.ylabel('Reales')
plt.title('Plano Z')
plt.legend(loc='upper left')

# Mostrar el grfico
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Distancia entre los centros de los crculos
l = np.sqrt((centro[0] - centro2[0])**2 + (centro[1] - centro2[1])**2)

# Calcular 'a'
a = l / np.sqrt(radio**4 - 2 * radio**2 * radio2**2 + radio2**4)

# Calcular radios de la transformacin a W
rho1 = (np.sqrt(1 + 4 * radio**2 * a**2) - 1) / (2 * radio * a**2)
rho2 = (np.sqrt(1 + 4 * radio2**2 * a**2) - 1) / (2 * radio2 * a**2)

# Generar puntos en el crculo
# Circulo 1
beta = np.linspace(0, 2*np.pi, 100)
x = rho1 * np.cos(beta)
y = rho1 * np.sin(beta)

# Circulo 2
beta2 = np.linspace(0, 2*np.pi, 100)
x2 = rho2 * np.cos(beta2)
y2 = rho2 * np.sin(beta2)

# Graficar los crculos en la misma figura
plt.figure(figsize=(6, 6))

# Circulo 1
plt.plot(x, y, label='Crculo 1')
plt.plot(0, 0, 'bo', label='Centro 1')

# Circulo 2
plt.plot(x2, y2, label='Crculo 2')
plt.plot(0, 0, 'ro', label='Centro 2')

# Configurar ejes y etiquetas
plt.axis('equal')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Plano W')
plt.legend(loc='upper left')

# Mostrar el grfico
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Definir la funcin potencial
def pot(rho, v1, rho1, rho2):
    return (v1 / math.log(rho2/rho1)) * math.log(rho/rho1)

v1 = 1
# Generar datos para el eje x
x_values = np.linspace(rho1, rho2, 100)  # 100 puntos entre rho1 y rho2

# Calcular los valores de la funcin potencial para cada punto en x
pot_values = [pot(x, v1, rho1, rho2) for x in x_values]

# Crear un mapa de calor
plt.figure(figsize=(8, 6))
plt.scatter(x_values, pot_values, c=pot_values, cmap='viridis', marker='o')
plt.colorbar(label='Valor de Potencial')
plt.xlabel('Radio (rho)')
plt.ylabel('Potencial')
plt.title('Mapa de Calor del Potencial')

# Mostrar el grfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Definimos un valor arbitrario para el potencial en el radio exterior
v1 = 1

# Definir la funcin potencial
def pot(r, v1, rho1, rho2):
    return (v1 / math.log(rho2/rho1)) * np.log(r/rho1)

# Generar datos en coordenadas cilndricas
theta = np.linspace(0, 2*np.pi, 100)  # ngulos
r_values = np.linspace(rho1, rho2, 100)  # Radios

# Crear malla en coordenadas cilndricas
r, theta = np.meshgrid(r_values, theta)

# Convertir coordenadas cilndricas a cartesianas
x = r * np.cos(theta)
y = r * np.sin(theta)

# Calcular los valores de la funcin potencial para cada punto en coordenadas cartesianas
pot_values = pot(r, v1, rho1, rho2)


# Crear un mapa de calor
plt.figure(figsize=(8, 8))
plt.pcolormesh(x, y, pot_values, cmap='viridis')
plt.colorbar(label='Valor de Potencial')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Mapa de Calor del Potencial en el plano W')


plt.axis('equal')
# Mostrar el grfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize

plt.figure(figsize=(6, 6))
r_values = np.linspace(rho1, rho2, 100)

# Normalize para mapear los colores al rango [0, 1]
norm = Normalize(vmin=rho1, vmax=rho2)

# Aplicar la transformada inversa para obtener los valores del centro y el radio,
# y graficar los crculos
for i in range(100):
    center = 5.3 - (a*((r_values[i])**2))/(1-(a**2)*((r_values[i])**2)) 
    radius = ((r_values[i]))/(1-(a**2)*((r_values[i])**2))
    
    theta = np.linspace(0, 2*np.pi, 100)
    x = center + radius * np.cos(theta)
    y = radius * np.sin(theta)
    
    color = plt.cm.viridis(norm(r_values[i]))
    
    plt.plot(x, y, color=color)



# Crear un mapa de calor
plt.axis('equal')
plt.xlabel('Reales')
plt.ylabel('Imaginarios')
plt.title('Mapa de Calor del Potencial en el plano Z')
plt.legend()

plt.grid(True)
plt.show()
```
